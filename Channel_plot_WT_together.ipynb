```python
#Libraries needed to run the tool
import numpy as np
import pandas as pd
import matplotlib
import matplotlib as mpl
import sys
##%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.axes_grid1 import make_axes_locatable, axes_size
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
sns.set(style='white')
from pylab import *
```

```python
df1 = pd.read_excel("mydata_WT_channel1_final.new.xlsx", names= ["midpoints", "means1", 'SD'])
df2 = pd.read_excel("mydata_WT_channel2_final.new.xlsx", names= ["midpoints", "means2", 'SD'])
df3 = pd.read_excel("mydata_WT_channel3_10ns_final.new.xlsx", names= ["midpoints", "means3", 'SD'])
```

```python
##plot of the data 
fig = plt.figure(1034, figsize=(17,11))
ax1 = fig.add_subplot(111)
from pylab import *
rc('axes', linewidth=5.5) 
#error = [df_whole['SD']]
ax1.set_ylabel('Pore radius ($\AA$)', fontsize=54, fontname='arial', labelpad=10)
ax1.set_xlabel('Pore coordinate $\zeta$ ($\AA$)', fontsize=54, fontname='arial', labelpad=10)
#ax1.ylim([0, 3.5])
plt.yticks(np.arange(2, 20.01, 2), fontname='arial')
ax1.plot(df2['midpoints'], df1["means1"], color='black', linewidth=4.0, label='CHANNEL 1')
ax1.plot(df2['midpoints'], df2["means2"], color='blue', linewidth=4.0, label='CHANNEL 2')
ax1.plot(df2['midpoints'], df3["means3"], color='green', linewidth=4.0, label='CHANNEL 3')
##ax1.plot(df4['midpoints'], df4["means4"], color='red', linewidth=4.0, label='40ns')
##ax1.plot(df5['midpoints'], df5["means5"], color='magenta', linewidth=4.0, label='50ns')
##ax1.plot(df6['midpoints'], df6["means6"], color='orange', linewidth=4.0, label='90ns')
##ax1.plot(df7['midpoints'], df7["means7"], color='cyan', linewidth=4.0, label='105ns')

##plt.errorbar(df1_D55N['midpoints'],df_whole['avg'], color='black', linewidth=4.0, yerr=df_whole['SD'], capsize = 5, ecolor='blue', elinewidth=3)
#ax1.fill_between(df1_D55N['midpoints'], df_whole['avg']-df_whole['SD'], df_whole['avg']+df_whole['SD'], color='r')
#ax1.plot(df2['midpoints'], df2["means2"], color='blue', linewidth=4.0, label='Hole-2')
#ax1.plot(df3['midpoints'], df3["means3"], color='green', linewidth=4.0, label='Hole-3')
ax1.set_xlim([80, -80])
ax1.set_ylim([0, 12])
#ax1.yticks(np.arange(0, 3.5, 0.5))
ax1.xaxis.set_minor_locator(plt.MultipleLocator(20))# set specify ticks label size
ax1.yaxis.set_minor_locator(plt.MultipleLocator(1))
ax1.tick_params(axis ='both', which= 'major', labelsize=52, direction='in', length=18, top=0, bottom=1, right=0, left=1, width=3.7)
#ax1.tick_params(axis ='both', which= 'major', left = 'bool', labelsize=20)
ax1.tick_params(axis ='both', which= 'minor', left='bool', labelsize=52, direction='in', length=10, width=3.7)
#ax1.tick_params(axis='both', which= 'major', bottom = 'bool', labelsize=20 )
#ax1.tick_params(axis='both', which= 'minor', bottom = 'bool', labelsize=20, direction='in', length=8)
ax1.legend(fontsize=25, bbox_to_anchor=(0.05, 0.40, 0.75, 0.58), edgecolor='white')
ax1.tick_params(axis='both', which='major', pad=15)
plt.title("Channel radii (WT)", loc = 'left', y=1.07, x=0.30, fontsize=56, fontname='arial', color='blue')
# Saving the plot as a 'png'
plt.savefig('All_channel_plot_together_WT.new.png',bbox_inches='tight', dpi=600)
##plt.savefig('channel_plot_together_WT.new.eps',format='eps', bbox_inches='tight', dpi=600)
plt.show()
```

```python
df_whole = pd.concat([df1["means1"], df2["means2"], df3["means3"]], axis=1)
```

```python
df_whole
```

```python
df_whole['SD'] = df_whole[['means1', 'means2', 'means3']].std(axis=1)
```

```python
df_whole['avg'] = df_whole[['means1', 'means2', 'means3']].mean(axis=1)
```

```python
data=pd.concat([df1["midpoints"], df_whole['avg'], df_whole['SD']], axis=1)
```

```python
##printing data into a excel sheet
writer = pd.ExcelWriter('mydata_WT_avg_channel_radius_new.xlsx')
##define the sheet name 'DataFrame'
data.to_excel(writer, 'DataFrame')
writer.save()
```
